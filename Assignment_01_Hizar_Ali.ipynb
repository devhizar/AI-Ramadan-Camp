```python
## Task 7: Marks Sheet with Grades
```

```python
total = 0
for i in range(5):
    marks = float(input(f"Enter marks for subject {i+1}: "))
    total += marks

average = total / 5

if average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 50:
    grade = "D"
else:
    grade = "F"

print("Total Marks:", total)
print("Average:", average)
print("Grade:", grade)
```
